```python
import networkx as nx
import math
import random
```

```python
with open('./data/com-youtube.ungraph.txt') as f:
    lines = f.readlines()
    
all_edges = [line.strip().split() for line in lines]
del all_edges[0:4] # to remove useless text of the file

all_edges
```

```python
G = nx.Graph()
G.add_edges_from(all_edges)
```

```python
V=len(G.nodes())
E=len(G.edges())
print 'Number of nodes:', V
print 'Number of edges:', E
print '-----------'
print 'We have {} connected component (CC)'.format(nx.number_connected_components(G))
print 'So the graph is connected.'
```

```python
from networkx.algorithms import community
communities_generator = community.girvan_newman(G)
```
